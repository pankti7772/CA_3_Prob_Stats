# Descriptive Statistics:

You can calculate descriptive statistics like mean, median, standard deviation, and skewness to understand the central tendency and variability of the data. The pandas and numpy libraries can be used for this.

```python
import statistics

# Calculate the mean
mean = statistics.mean(df['Transaction Amount'])

# Calculate the median
median = statistics.median(df['Transaction Amount'])

# Calculate the mode (if it exists)
try:
    mode = statistics.mode(df['Transaction Amount'])
except statistics.StatisticsError:
    mode = "No unique mode found"

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
```

```python
# Import statistics module
import statistics

# Calculate the standard deviation of Transaction Amount
std_dev = statistics.stdev(df['Transaction Amount'])

# Print the standard deviation
print(f"Standard deviation: {std_dev}")
```

# Statistical Tests:

You can use statistical tests such as the Kolmogorov-Smirnov test to check if the data follows a specific distribution. The scipy library can be used for this purpose.

```python
from scipy.stats import kstest, norm

column_to_test = df['Transaction Amount']

# Parameters for the normal distribution
mean = column_to_test.mean()
std_deviation = column_to_test.std()

# Test for normal distribution
ks_statistic, ks_p_value = kstest(column_to_test, 'norm', args=(mean, std_deviation))
if ks_p_value < 0.05:
    print("The data does not follow a normal distribution.")
else:
    print("The data follows a normal distribution.")
```

# Stratified Sampling:
In stratified sampling, you divide your dataset into subgroups or strata based on a specific characteristic, and then you sample proportionally from each stratum.

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Given data
data = {
    'Transaction Date and Time': ['2022-09-24 13:54:27', '2023-03-18 01:05:36', '2021-12-16 06:22:24'],
    'Transaction Amount': [285.88, 3939.01, 1687.33],
    'Cardholder Name': ['Shray Soman', 'Riya Bobal', 'Mannat Rout'],
    'Merchant Category Code (MCC)': ['3590', '9297', '2940'],
    'Card Expiration Date': ['04/29', '11/29', '03/31'],
    'Transaction ID': ['b7f69cbc-a03d-41f8-adca-75920b0242c3', 'f09cecd6-0d74-4551-8089-a31d64e6b9c5',
                       '6b4e4e43-5b73-4906-9973-299a1b2a5e71'],
    'Fraud Flag or Label': ['1', '1', '0'],  # Assuming 0 represents Non-Fraud
    'Previous Transactions': ['None', '3 or more', '2'],
    'Transaction Source': ['Online', 'In-Person', 'Online'],
    'IP Address': ['18.106.240.6', '17.190.112.46', '196.153.28.131'],
    'Device Information': ['Tablet', 'Mobile', 'Desktop'],
    'User Account Information': ['NaN', 'NaN', 'NaN'],
    'Transaction Notes': ['Approved', 'Declined', 'Approved'],
}

# Convert the data to a pandas DataFrame
df = pd.DataFrame(data)

# Identify strata based on 'Transaction Source'
strata = df.groupby('Transaction Source')

# Perform stratified sampling
stratified_sampled_data = strata.apply(lambda x: x.sample(frac=0.8))  # Adjust the fraction as needed

# Display the stratified sampled data
print("Stratified Sampled Data:")
print(stratified_sampled_data)
```

# Systematic Sampling: 
Systematic sampling involves selecting every "k"th data point from the dataset, where "k" is the sampling interval.

```python
df = pd.DataFrame(data)

# Sampling interval
interval = 1  # You can change this to any positive integer

# Perform systematic sampling
sample = df.iloc[::interval]
print(sample)
```

# Cluster Sampling:
In cluster sampling, you divide the dataset into clusters and randomly select a subset of clusters to sample. Then, you include all data points within the selected clusters.

```python
import random

# Convert the data to a pandas DataFrame
df = pd.DataFrame(data)

# Add a 'Cluster' column (assuming each row is a separate cluster)
df['Cluster'] = range(len(df))

# Number of clusters to select
num_selected_clusters = 2

# Randomly select clusters
selected_clusters = random.sample(list(df['Cluster']), num_selected_clusters)

# Include all data points within the selected clusters
sampled_data = df[df['Cluster'].isin(selected_clusters)]

print(sampled_data)
```

# Bootstrapping: 
Bootstrapping involves creating multiple samples with replacement from the original dataset. It's often used for estimating statistics, such as the mean or standard deviation, and for building confidence intervals.

```python
import numpy as np

# Convert the data to a 2D array
data_array = np.array(list(data.values()))

# Flatten the 2D array to a 1D array
flattened_data = data_array.flatten()

# Set a random seed for reproducibility
np.random.seed(42)

# Generate a bootstrapped sample
bootstrap_sample = np.random.choice(flattened_data, size=len(flattened_data), replace=True)

print(bootstrap_sample)
```
